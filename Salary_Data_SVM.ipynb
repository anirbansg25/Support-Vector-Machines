```python
# Feature Selection phase to check for importance of different features
```

```python
# Split the data into X & y

X = d4.drop('High_Salary', axis = 1).values
y = d4['High_Salary']

y = y.astype(int)

print(X.shape)
print(y.shape)
```

```python
# Run a Tree-based estimators (Decision Tree in this case)
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=15, criterion = 'entropy', max_depth = 10)
dt.fit(X,y)
```

```python
# Running Feature Importance
fi_col = []
fi = []

for i,column in enumerate(d4.drop('High_Salary', axis = 1)):
    print('The feature importance for {} is : {}'.format(column, dt.feature_importances_[i]))
    
    fi_col.append(column)
    fi.append(dt.feature_importances_[i])
```

```python
# Creating a Dataframe
fi_col
fi

fi_df = zip(fi_col, fi)
fi_df = pd.DataFrame(fi_df, columns = ['Feature','Feature Importance'])


# Ordering the data
fi_df = fi_df.sort_values('Feature Importance', ascending = False).reset_index()
fi_df
```

```python
# Since all the features have some importance as they are non-zero, we won't remove any particular feature
```
